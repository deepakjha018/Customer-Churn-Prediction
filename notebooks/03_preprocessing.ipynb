# Milestone 4: Data Preprocessing

## Objective

The goal of preprocessing is to transform the raw dataset into a clean and machine-learning-ready format.

The preprocessing steps include:

- Removing unnecessary features
- Handling missing values
- Correcting data types
- Encoding categorical variables
- Splitting the dataset
- Scaling numerical features
- Handling class imbalance using SMOTE

# Preprocessing Strategy

# Preprocessing Summary

The preprocessing pipeline prepared the dataset for machine learning by:

- Removing the identifier column (`customerID`)

- Removing rows with missing values
- Encoding the target variable (`Churn`)
- Splitting the dataset into training and testing sets using stratification
- Applying a `ColumnTransformer` to preprocess numerical and categorical features
- Balancing the training data using SMOTE

The dataset is now ready for model training.
